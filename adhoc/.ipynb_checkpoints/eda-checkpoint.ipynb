```python
import json
import re
from glob import glob
from tqdm.notebook import tqdm

import nglview
from pymatgen.core import Structure
```

```python
structure_paths = glob('../data/dichalcogenides_public/structures/*.json')
structures = {
    list(filter(lambda x: x != '', re.split(r'\/(\w+)\.json', path)))[-1]:
    Structure.from_dict(json.load(open(path, 'r')))
    for path in tqdm(structure_paths)
}
```

```python
type(structures[list(structures.keys())[1]])
```

```python
example_molecule = structures[list(structures.keys())[1]]
example_df = example_molecule.as_dataframe()
```
